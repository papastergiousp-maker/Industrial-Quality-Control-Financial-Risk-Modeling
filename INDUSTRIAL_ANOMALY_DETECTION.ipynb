# 🏭 Industrial Quality Control & Financial Risk Modeling

## Business Problem

In food production, pH deviations during manufacturing directly translate
to financial losses — either through correction costs (additives, rework)
or critical waste (batch disposal). The business question is:

> **What is the total financial impact of pH-based batch failures across
> a 60-day production cycle, and how are losses distributed between
> correctable and critical batches?**

---

## Data & Methodology

**Real production pH measurements from D. Nomikos Food Industry (Φάρσαλα)
cannot be shared due to data confidentiality agreements.**

This model uses a statistically calibrated simulation based on observed
production parameters during my Quality Control internship (June–September 2024):

| Parameter | Value | Source |
|:----------|:------|:-------|
| Distribution | Normal | pH measurements follow Gaussian distribution in controlled production |
| Mean (loc) | 4.22 | Calibrated to observed production baseline |
| Std Dev (scale) | 0.12 | Reflects real batch-to-batch variability |
| Sample size | 480 | 8 measurements/day × 60 days |

**Business Logic (pH Thresholds):**

| pH Range | Classification | Financial Impact |
|:---------|:---------------|:----------------|
| pH ≤ 4.30 | ✅ Optimal | €0 |
| 4.30 < pH ≤ 4.50 | ⚠️ Correction (Citric Acid) | €500/batch |
| pH > 4.50 | ❌ Critical Waste | €5,000/batch |

## Tools & Libraries

`Python` `pandas` `NumPy` `Matplotlib` `Seaborn`

## Output

This notebook produces a **Production Quality Financial Report** —
a cost summary by batch classification, not just a visualisation.

---
## Step 1 — Imports & Business Logic

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Business Logic: pH thresholds define batch classification and financial cost
# Based on HACCP / ISO 22000 production parameters — D. Nomikos Food Industry
def PH_ANALYSIS(ph):
    if ph <= 4.30:
        return "Optimal", 0
    elif ph <= 4.50:
        return "Correction (Citric Acid)", 500
    else:
        return "Critical Waste", 5000
```

---
## Step 2 — Simulate Production Data

480 batches = 8 measurements/day × 60 days.  
Parameters calibrated to observed production baseline (confidential data, D. Nomikos 2024).

```python
np.random.seed(42)
random_ph_values = np.random.normal(loc=4.22, scale=0.12, size=480)

# Build DataFrame with classification and financial loss per batch
results = []
for ph in random_ph_values:
    status, loss = PH_ANALYSIS(ph)
    results.append({'pH': ph, 'Status': status, 'Loss': loss})

df = pd.DataFrame(results)
print(f"Dataset created: {len(df)} batches")
print(df['Status'].value_counts())
```

---
## Step 3 — Financial Impact Calculation

```python
waste_costs       = df[df['Status'] == "Critical Waste"]['Loss'].sum()
correction_costs  = df[df['Status'] == "Correction (Citric Acid)"]['Loss'].sum()
total_financial_impact = df['Loss'].sum()

# Production Quality Financial Report
print("=" * 40)
print("   PRODUCTION QUALITY FINANCIAL REPORT")
print("=" * 40)
print(f"Total Batches Analyzed:       {len(df)}")
print(f"Cost of Corrections (Acid):   {correction_costs}€")
print(f"Cost of Critical Waste:       {waste_costs}€")
print("-" * 40)
print(f"TOTAL FINANCIAL IMPACT:       {total_financial_impact}€")
print("=" * 40)
```

---
## Step 4 — Quality Control Distribution Chart

Bar chart showing batch count per classification category,
with total financial impact in the title.

```python
palette = {
    "Optimal": "#2ecc71",
    "Correction (Citric Acid)": "#f39c12",
    "Critical Waste": "#e74c3c"
}

ax = sns.countplot(
    x='Status',
    data=df,
    hue='Status',
    palette=palette,
    order=['Optimal', 'Correction (Citric Acid)', 'Critical Waste'],
    legend=False
)

# Labels on top of each bar
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center',
        xytext=(0, 9),
        textcoords='offset points',
        fontsize=11,
        fontweight='bold'
    )

plt.title(f'Quality Control Distribution | Total Impact: {total_financial_impact}€', fontsize=14)
plt.xlabel('Batch Status')
plt.ylabel('Number of Batches')
plt.ylim(0, df['Status'].value_counts().max() + 50)

plt.tight_layout()
plt.show()
```

---
## Step 5 — pH Distribution Histogram

Visualises the full pH distribution across all 480 batches,
with threshold lines showing the Optimal / Correction / Critical Waste boundaries.

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['pH'], bins=30, kde=True, color='#3498db', alpha=0.7)

# Threshold lines
plt.axvline(x=4.30, color='#f39c12', linestyle='--', linewidth=1.5, label='Threshold: Correction (pH 4.30)')
plt.axvline(x=4.50, color='#e74c3c', linestyle='--', linewidth=1.5, label='Threshold: Critical Waste (pH 4.50)')

plt.title('pH Distribution — 480 Production Batches', fontsize=14)
plt.xlabel('pH Value')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

---
## Step 6 — Financial Loss Breakdown

Pie chart showing the split of total financial loss
between correction costs and critical waste.

```python
labels  = ['Correction (Citric Acid)', 'Critical Waste']
values  = [correction_costs, waste_costs]
colors  = ['#f39c12', '#e74c3c']

plt.figure(figsize=(7, 5))
plt.pie(
    values,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    textprops={'fontsize': 12}
)
plt.title(f'Financial Loss Breakdown | Total: {total_financial_impact}€', fontsize=13)
plt.tight_layout()
plt.show()
```

## Key Findings

| Metric | Value |
|:-------|:------|
| Total batches analyzed | 480 |
| Optimal batches | 364 (75.8%) |
| Correction batches | 112 (23.3%) |
| Critical Waste batches | 4 (0.8%) |

**Insight:** Although Critical Waste batches are rare (0.8%), each costs
